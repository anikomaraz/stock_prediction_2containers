```python
import datetime
import pandas as pd
```

```python
# connect to google 

from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360) 
```

```python
def get_google_search(company_ticker):
    
    # build payload
    kw_list = [company_ticker]
    
    pytrends.build_payload(kw_list, cat=0, timeframe='2022-05-28 2022-10-31') 
    second_half = pytrends.interest_over_time() 
    second_half = second_half.reset_index() 
    
    pytrends.build_payload(kw_list, cat=0, timeframe='2022-01-01 2022-05-27') 
    first_half = pytrends.interest_over_time() 
    first_half = first_half.reset_index() 
    
    data = pd.concat([first_half,second_half]).drop(['isPartial'], axis=1)
    
    return data
```

```python
get_google_search('AAPL')
```

```python
# build payload

kw_list = ["AAPL"] # list of keywords to get data 

pytrends.build_payload(kw_list, cat=0, timeframe='2022-05-28 2022-10-31') 
```
